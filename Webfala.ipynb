```python
#union() – Combines Two Sets

set1 = {1, 2, 3}
set2 = {3, 4, 5}
result = set2.union(set1)
print(result)
```

```python
# Alternatively, use the | operator:

result = set1 | set2
print(result)
```

```python
#intersection() – Common Elements Between Sets

set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
result = set2.intersection(set1)
print(result)
```

```python
# Alternatively, use the & operator:

result = set1 & set2
print(result)
```

```python
#difference() – Elements in One Set but Not the Other
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
result = set2.difference(set1)
print(result)

#or
result = set1.difference(set2)
print(result)
```

```python
# Alternatively, use the - operator:

result = set1 - set2
print(result)
```

```python
#symmetric_difference() – Elements in Either Set but Not Both

set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
result = set1.symmetric_difference(set2)
print(result)
```

```python
# Alternatively, use the ^ operator:

result = set1 ^ set2
print(result)
```

Generating Random Numbers
(random.random()) generates a random decimal between 0.0 and 1.0.

```python
#Random Numbers (random.random()) generates a random decimal between 0.0 and 1.0.

random_num = random.random()
print(random_num)
```

```python
#Random Integer in a Range (random.randint(a, b)) - Generates a random integer between a and b (inclusive).

random_num = random.randint(1, 10)
print(random_num)
```

```python
#Random Float in a Range (random.uniform(a, b)) - Generates a random floating-point number in the given range.

random_num = random.uniform(1, 5)
random_num
```

```python
#Random Choice from a List - Select a Random Element (random.choice())

names = ['Abiola', 'Halimah', 'Zainab', 'Esther', 'Lola', 'Carina', 'Salamat']
random_name = random.choice(names)
print(random_name)
```

```python
# Select Multiple Random Elements (random.sample())

two_random_names = random.sample(names, 2)
print(two_random_names)
```
